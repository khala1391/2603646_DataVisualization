## Regression

```python
alt.Chart(df).mark_circle().encode(
    x='education',
    y='prestige',
    color='type:N',
    tooltip='income'
)
```

```python
chart = alt.Chart(df).mark_circle().encode(
    x= alt.X('education',scale=alt.Scale(domain=[0, 20])),
    y='prestige',
    color='type',
    tooltip=['income'],
)
chart+chart.transform_regression('education','prestige').mark_line()
```

```python
educationPlot = alt.Chart(df).mark_circle().encode(
    x= alt.X('education',scale=alt.Scale(domain=[0, 20])),
    y='prestige',
    color='type',
)
incomePlot = alt.Chart(df).mark_circle().encode(
    x='income',
    y='prestige',
    color='type',
)

educationPlot | incomePlot
```
